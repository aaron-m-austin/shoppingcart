```python
from typing import Dict

class Product:
    """Adding variable type for these?"""
    def __init__(self, name: str, price: float, stock: int):
        self.name = name
        self.price = float(price)
        self.stock = int(stock)


class ShoppingCart:
    
    def __init__(self):
        self.items: Dict[str, Dict[str, object]] = {}

    def add_product(self, product: Product, quantity: int) -> bool:
        # validate quantity
        if quantity <= 0:
            print("Quantity must be positive.")
            return False

        # current quantity in the cart
        current_qty = self.items.get(product.name, {}).get("qty", 0)
        new_total_qty = current_qty + int(quantity)

        # check for item availability
        if new_total_qty > product.stock:
            print(f"Cannot add {quantity} x {product.name}: only {product.stock - current_qty} available (stock limit).")
            return False

        # add or update product in the cart
        if product.name in self.items:
            self.items[product.name]["qty"] = new_total_qty
        else:
            self.items[product.name] = {"product": product, "qty": int(quantity)}

        print(f"Added {quantity} x {product.name} to cart. (Now {new_total_qty} in cart.)")
        return True

    def remove_product(self, product_name: str) -> bool:
        if product_name in self.items:
            del self.items[product_name]
            print(f"Removed {product_name} from cart.")
            return True
        else:
            print(f"{product_name} not found in cart.")
            return False
    apple = Product("Apple", 1.00, 10)
    cart = ShoppingCart()

    cart.add_product(apple, 5)
    cart.add_product(apple, 7)  # Should fail due to stock limit
    cart.remove_product("Apple")

```
